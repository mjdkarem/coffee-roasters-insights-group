```python
#CLEANING DATA
# Remove exact duplicate rows
df = df.drop_duplicates()
```

```python
#HANLING MISSING VALUES
print(df.isnull().sum())
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna('unknown')
```

```python
# Strip leading/trailing spaces and convert to lowercase
df = df.apply(lambda x: x.str.strip().str.lower() if x.dtype == "object" else x)
```

```python
# Fix Data Types
df['Grading Date'] = pd.to_datetime(df['Grading Date'], errors='coerce')
df['Expiration'] = pd.to_datetime(df['Expiration'], errors='coerce')

# Remove 'kg', strip spaces, and convert to numeric
df['Bag Weight'] = df['Bag Weight'].astype(str).str.replace('kg', '', case=False, regex=True).str.strip()

# Convert cleaned string to numeric
df['Bag Weight'] = pd.to_numeric(df['Bag Weight'], errors='coerce').astype('Int64')

# Apply title case to all object or category columns
# Apply title case to all object columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype(str).str.title()

for col in df.select_dtypes(include=['object', 'category']).columns:
    df[col] = df[col].str.title()

categorical_columns = [
    'Country of Origin', 'Region', 'Processing Method',
    'Variety', 'Status', 'Color', 'Certification Body'
]
for col in categorical_columns:
    df[col] = df[col].astype('category')
```

```python
df = df.reset_index(drop=True)
```
